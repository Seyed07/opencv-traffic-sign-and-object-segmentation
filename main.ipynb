```python
# ====== Traffic Sign Classifier (OpenCV-only)  stricter circle, AR-based rectangles ======
import math, os, cv2 as cv, numpy as np
from google.colab import files
from google.colab.patches import cv2_imshow

CATEGORIES_FA = {"Warning":"","Prohibitory":"","Informational":"","Guide":" "}

# ---------- Color masks ----------
def color_masks(img):
    hsv = cv.cvtColor(img, cv.COLOR_BGR2HSV)
    # Red (two ranges)
    red = cv.bitwise_or(
        cv.inRange(hsv, np.array([0,80,70]),   np.array([10,255,255])),
        cv.inRange(hsv, np.array([170,80,70]), np.array([180,255,255]))
    )
    # Blue
    blue = cv.inRange(hsv, np.array([95,70,60]), np.array([135,255,255]))
    k = cv.getStructuringElement(cv.MORPH_ELLIPSE,(5,5))
    red  = cv.morphologyEx(red,  cv.MORPH_OPEN, k, 1); red  = cv.morphologyEx(red,  cv.MORPH_CLOSE, k, 2)
    blue = cv.morphologyEx(blue, cv.MORPH_OPEN, k, 1); blue = cv.morphologyEx(blue, cv.MORPH_CLOSE, k, 2)
    return red, blue

def biggest_contour(mask):
    cnts,_ = cv.findContours(mask, cv.RETR_EXTERNAL, cv.CHAIN_APPROX_SIMPLE)
    return max(cnts, key=cv.contourArea) if cnts else None

# ---------- Features ----------
def poly_features(cnt):
    peri  = cv.arcLength(cnt, True)
    area  = max(cv.contourArea(cnt), 1.0)
    approx = cv.approxPolyDP(cnt, 0.02*peri, True)
    v = len(approx)
    # rotated rectangle (rotation-invariant)
    (cx,cy),(rw,rh),theta = cv.minAreaRect(cnt)
    rw, rh = float(rw), float(rh)
    if rw < rh: rw, rh = rh, rw
    ar_rot = rw/max(rh,1e-6)                 # >= 1
    rect_area = max(rw*rh, 1e-6)
    rect_fill = area/rect_area               # 1 for perfectly filled rectangle, ~0.785 for perfect circle
    circularity = 4.0*math.pi*area/(peri*peri + 1e-6)  # 1 for circle
    return dict(v=v, ar_rot=ar_rot, rect_fill=rect_fill, circ=circularity, approx=approx,
                peri=peri, area=area)

# ---------- Hough circle (optional vote) ----------
def detect_circle_hough(gray):
    g = cv.GaussianBlur(gray,(7,7),1.5)
    circles = cv.HoughCircles(g, cv.HOUGH_GRADIENT, dp=1.2, minDist=gray.shape[0]//4,
                              param1=120, param2=45, minRadius=0, maxRadius=0)
    return None if circles is None else np.uint16(np.around(circles[0]))

# ---------- Circle template / triangle template (Hu moments) ----------
CIRCLE_TMPL = None
def get_circle_template():
    global CIRCLE_TMPL
    if CIRCLE_TMPL is not None: return CIRCLE_TMPL
    img = np.zeros((400,400), np.uint8)
    cv.circle(img, (200,200), 120, 255, thickness=4)
    cnts,_ = cv.findContours(img, cv.RETR_EXTERNAL, cv.CHAIN_APPROX_NONE)
    CIRCLE_TMPL = cnts[0]
    return CIRCLE_TMPL

TRI_TMPL = np.array([[200,60],[80,300],[320,300]], np.int32).reshape(-1,1,2)

def is_triangle_like(cnt, thresh=0.08):
    s = cv.matchShapes(cnt, TRI_TMPL, cv.CONTOURS_MATCH_I1, 0.0)
    return s < thresh

# ---------- Strict circle check (multi-criterion voting) ----------
def is_circle_strict(cnt, feat):
    # 1) circularity strong
    cond_circ = feat["circ"] >= 0.86

    # 2) near-square bounding box
    cond_box  = feat["ar_rot"] <= 1.20

    # 3) circle shouldn't fill a rectangle too well (rect_fill ~ /4  0.785)
    cond_rectfill = feat["rect_fill"] <= 0.83

    # 4) area close to minEnclosingCircle area
    (x,y), r = cv.minEnclosingCircle(cnt)
    enc_area = math.pi * (float(r)**2)
    cond_enc = (feat["area"] / max(enc_area,1e-6)) >= 0.80

    # 5) radius stability: std(radius)/mean(radius) must be small
    pts = cnt.reshape(-1,2).astype(np.float32)
    dists = np.linalg.norm(pts - np.array([[x,y]], dtype=np.float32), axis=1)
    rad_cv = np.std(dists) / max(np.mean(dists), 1e-6)
    cond_rad = rad_cv <= 0.08

    # 6) Hu match to circle template
    circle_cnt = get_circle_template()
    hu = cv.matchShapes(cnt, circle_cnt, cv.CONTOURS_MATCH_I1, 0.0)
    cond_hu = hu <= 0.03

    # 7) Hough circle as an extra vote
    cond_hough = False  # filled by caller if desired

    votes = [cond_circ, cond_box, cond_rectfill, cond_enc, cond_rad, cond_hu, cond_hough]
    return sum(votes) >= 4, dict(circ=cond_circ, box=cond_box, rectfill=cond_rectfill,
                                 enc=cond_enc, rad=cond_rad, hu=cond_hu)

# ---------- Classifier (with your color rules) ----------
GUIDE_AR = 1.55         # wide if >= this (tune 1.501.65)
INFO_AR_MAX = 1.45      # info near-square if <= this
RECT_FILL_MIN = 0.70    # must look like a rectangle

def classify(img):
    h,w = img.shape[:2]
    red, blue = color_masks(img)
    red_area  = cv.countNonZero(red)
    blue_area = cv.countNonZero(blue)
    area_thresh = 0.003*h*w

    # Build a contour from the dominant mask, else edges
    mask = red if red_area >= blue_area else blue
    cnt = biggest_contour(mask)
    if cnt is None or cv.contourArea(cnt) < 0.004*h*w:
        edges = cv.Canny(cv.GaussianBlur(cv.cvtColor(img, cv.COLOR_BGR2GRAY),(5,5),0), 60, 180)
        cnts,_ = cv.findContours(edges, cv.RETR_EXTERNAL, cv.CHAIN_APPROX_SIMPLE)
        cnt = max(cnts, key=cv.contourArea) if cnts else None
    if cnt is None:
        return {"label":"UNKNOWN","reason":"no contour"}

    feat = poly_features(cnt)

    # (1) Prohibitory: circle only if strict vote passes
    gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
    hough_found = detect_circle_hough(gray) is not None
    is_circle, votes = is_circle_strict(cnt, feat)
    if is_circle and hough_found:
        return {"label":"Prohibitory","rule":"Strict circle + Hough","cnt":cnt}
    # if too strict, allow strict circle alone (uncomment next 2 lines):
    # if is_circle and feat["circ"] >= 0.90:
    #     return {"label":"Prohibitory","rule":"Strict circle (high circularity)","cnt":cnt}

    # (2) Warning: triangle but require RED dominance
    if red_area > max(blue_area, area_thresh) and (feat["v"]==3 or is_triangle_like(cnt)):
        return {"label":"Warning","rule":"Triangle + red-dominant","cnt":cnt}

    # (3) Rectangle family
    if feat["rect_fill"] >= RECT_FILL_MIN:
        # Informational = BLUE and not wide
        if blue_area > max(red_area, area_thresh) and feat["ar_rot"] <= INFO_AR_MAX:
            return {"label":"Informational","rule":f"Blue rectangle, AR={feat['ar_rot']:.2f}  {INFO_AR_MAX}","cnt":cnt}
        # Guide = wide rectangle (color-agnostic)
        if feat["ar_rot"] >= GUIDE_AR:
            return {"label":"Guide","rule":f"Wide rectangle, AR={feat['ar_rot']:.2f}  {GUIDE_AR}","cnt":cnt}

    # Fallback for 4-vertex rectangles
    if feat["v"]==4:
        return {"label": "Guide" if feat["ar_rot"]>=GUIDE_AR else "Informational",
                "rule": f"v=4 fallback, AR={feat['ar_rot']:.2f}","cnt":cnt}

    return {"label":"UNKNOWN","reason":"no rule matched","cnt":cnt}

# ---------- Annotate ----------
def annotate(img, cnt, label, rule):
    vis = img.copy()
    if cnt is not None:
        cv.drawContours(vis,[cnt],-1,(0,255,0),3)
        x,y,w,h = cv.boundingRect(cnt)
        cv.rectangle(vis,(x,y),(x+w,y+h),(0,0,255),2)
        banner = f"{label} / {CATEGORIES_FA.get(label,label)}"
        cv.rectangle(vis,(x,max(0,y-28)),(x+420,y),(0,0,0),-1)
        cv.putText(vis,banner,(x+6,y-8),cv.FONT_HERSHEY_SIMPLEX,0.7,(255,255,255),2,cv.LINE_AA)
        if rule:
            cv.putText(vis,rule,(10,img.shape[0]-10),cv.FONT_HERSHEY_SIMPLEX,0.5,(0,0,0),2,cv.LINE_AA)
            cv.putText(vis,rule,(10,img.shape[0]-10),cv.FONT_HERSHEY_SIMPLEX,0.5,(255,255,255),1,cv.LINE_AA)
    return vis

def process_image(path):
    img = cv.imread(str(path), cv.IMREAD_COLOR)
    if img is None:
        return {"path":str(path),"error":"cannot read image"}
    res = classify(img)
    vis = annotate(img, res.get("cnt"), res.get("label","UNKNOWN"), res.get("rule",""))
    res.update({"path":str(path),"annotated":vis})
    return res

# ---------- Upload loop ----------
os.makedirs("/content/outputs_strict", exist_ok=True)
print("Upload images; results will show and save to /content/outputs_strict.")

while True:
    up = files.upload()
    if not up:
        print("No file uploaded. Exiting.")
        break
    for name,data in up.items():
        p = f"/content/{name}"
        with open(p,"wb") as f: f.write(data)
        r = process_image(p)
        outp = f"/content/outputs_strict/{name}"
        cv.imwrite(outp, r["annotated"])
        print(f"{name} -> {r['label']} | {r.get('rule','')} | saved: {outp}")
        cv2_imshow(r["annotated"])
    if input("Again? (y/n): ").strip().lower() != "y":
        print("Done."); break
```

```python
import os, math, csv, pathlib
import cv2 as cv
import numpy as np
from google.colab import files
from google.colab.patches import cv2_imshow

CATEGORIES = ["Warning","Prohibitory","Informational"]
PERSIAN = {"Warning":"","Prohibitory":"","Informational":""}

# ---------------- Color preprocessing ----------------
def color_masks(img):
    hsv = cv.cvtColor(img, cv.COLOR_BGR2HSV)
    # Red: two hue spans
    red = cv.bitwise_or(
        cv.inRange(hsv, np.array([0,80,70]),   np.array([10,255,255])),
        cv.inRange(hsv, np.array([170,80,70]), np.array([180,255,255]))
    )
    # Blue
    blue = cv.inRange(hsv, np.array([95,70,60]), np.array([135,255,255]))
    k = cv.getStructuringElement(cv.MORPH_ELLIPSE,(5,5))
    red  = cv.morphologyEx(red,  cv.MORPH_OPEN, k, 1); red  = cv.morphologyEx(red,  cv.MORPH_CLOSE, k, 2)
    blue = cv.morphologyEx(blue, cv.MORPH_OPEN, k, 1); blue = cv.morphologyEx(blue, cv.MORPH_CLOSE, k, 2)
    return red, blue

# ---------------- Candidate extraction ----------------
def contours_from_mask(mask, min_area):
    cnts,_ = cv.findContours(mask, cv.RETR_EXTERNAL, cv.CHAIN_APPROX_SIMPLE)
    boxes = []
    for c in cnts:
        a = cv.contourArea(c)
        if a >= min_area:
            x,y,w,h = cv.boundingRect(c)
            boxes.append((x,y,w,h))
    return boxes

def iou(b1, b2):
    x1,y1,w1,h1 = b1; x2,y2,w2,h2 = b2
    xa, ya = max(x1,x2), max(y1,y2)
    xb, yb = min(x1+w1, x2+w2), min(y1+h1, y2+h2)
    inter = max(0, xb-xa) * max(0, yb-ya)
    if inter == 0: return 0.0
    a1, a2 = w1*h1, w2*h2
    return inter / float(a1 + a2 - inter)

def merge_boxes(boxes, iou_thr=0.2):
    boxes = sorted(boxes, key=lambda b: b[2]*b[3], reverse=True)
    kept = []
    for b in boxes:
        if all(iou(b, k) < iou_thr for k in kept):
            kept.append(b)
    return kept

def pad_box(b, img_shape, pad=8):
    x,y,w,h = b; H,W = img_shape[:2]
    x2, y2 = x+w, y+h
    return (max(0,x-pad), max(0,y-pad),
            min(W, x2+pad)-max(0,x-pad),
            min(H, y2+pad)-max(0,y-pad))

# ---------------- ROI features ----------------
def largest_cnt(bin_img):
    cnts,_ = cv.findContours(bin_img, cv.RETR_EXTERNAL, cv.CHAIN_APPROX_SIMPLE)
    return max(cnts, key=cv.contourArea) if cnts else None

def roi_contour(roi):
    gray = cv.cvtColor(roi, cv.COLOR_BGR2GRAY)
    edges = cv.Canny(cv.GaussianBlur(gray,(5,5),0), 60, 180)
    k = cv.getStructuringElement(cv.MORPH_ELLIPSE,(5,5))
    edges = cv.morphologyEx(edges, cv.MORPH_CLOSE, k, 2)
    return largest_cnt(edges)

def poly_features(cnt):
    peri  = cv.arcLength(cnt, True)
    area  = max(cv.contourArea(cnt), 1.0)
    approx = cv.approxPolyDP(cnt, 0.02*peri, True)
    v = len(approx)
    (cx,cy),(rw,rh),theta = cv.minAreaRect(cnt)
    rw, rh = float(rw), float(rh)
    if rw < rh: rw, rh = rh, rw
    ar_rot = rw/max(rh,1e-6)                 # >= 1
    rect_area = max(rw*rh, 1e-6)
    rect_fill = area/rect_area               # circle ~0.78, rectangle~1 (minus border)
    circularity = 4.0*math.pi*area/(peri*peri + 1e-6)  # circle -> 1
    return dict(v=v, ar_rot=ar_rot, rect_fill=rect_fill, circ=circularity, area=area, approx=approx)

# strict circle voting (no deep learning)
_CIRCLE_TMPL = None
def circle_template():
    global _CIRCLE_TMPL
    if _CIRCLE_TMPL is None:
        img = np.zeros((400,400), np.uint8)
        cv.circle(img, (200,200), 120, 255, 4)
        cnts,_ = cv.findContours(img, cv.RETR_EXTERNAL, cv.CHAIN_APPROX_NONE)
        _CIRCLE_TMPL = cnts[0]
    return _CIRCLE_TMPL

def is_circle_strict(cnt, feat):
    (x,y), r = cv.minEnclosingCircle(cnt)
    enc_area = math.pi * (float(r)**2)
    pts = cnt.reshape(-1,2).astype(np.float32)
    dists = np.linalg.norm(pts - np.array([[x,y]], dtype=np.float32), axis=1)
    rad_cv = np.std(dists) / max(np.mean(dists), 1e-6)
    hu = cv.matchShapes(cnt, circle_template(), cv.CONTOURS_MATCH_I1, 0.0)

    cond_circ      = feat["circ"] >= 0.86
    cond_box       = feat["ar_rot"] <= 1.20
    cond_rectfill  = feat["rect_fill"] <= 0.83
    cond_enc       = (feat["area"]/max(enc_area,1e-6)) >= 0.80
    cond_rad       = rad_cv <= 0.08
    cond_hu        = hu <= 0.03
    votes = [cond_circ, cond_box, cond_rectfill, cond_enc, cond_rad, cond_hu]
    return sum(votes) >= 4

# triangle template
TRI = np.array([[200,60],[80,300],[320,300]], np.int32).reshape(-1,1,2)
def is_triangle_like(cnt, thresh=0.08):
    return cv.matchShapes(cnt, TRI, cv.CONTOURS_MATCH_I1, 0.0) < thresh

# ---------------- Classify one ROI ----------------
RECT_FILL_MIN = 0.70
AREA_MIN_FRAC = 0.003     # ignore tiny color blobs (relative to full image)

def classify_roi(roi, full_masks):
    red_full, blue_full = full_masks
    # color dominance measured within ROI
    x0,y0,w,h = 0,0,roi.shape[1], roi.shape[0]  # for clarity
    # extract roi masks via slicing of full masks later (caller supplies crop)
    # but here we just recompute on ROI for simplicity
    red_roi, blue_roi = color_masks(roi)
    red_area  = cv.countNonZero(red_roi)
    blue_area = cv.countNonZero(blue_roi)
    roi_area  = roi.shape[0] * roi.shape[1]

    cnt = roi_contour(roi)
    if cnt is None:
        return "UNKNOWN", "no contour"

    feat = poly_features(cnt)

    # Prohibitory: strict circle + red presence
    if is_circle_strict(cnt, feat) and red_area > blue_area and red_area > AREA_MIN_FRAC*roi_area:
        return "Prohibitory", "strict circle + red-dominant"

    # Warning: triangle + red-dominant
    if (feat["v"]==3 or is_triangle_like(cnt)) and red_area > blue_area and red_area > AREA_MIN_FRAC*roi_area:
        return "Warning", "triangle + red-dominant"

    # Informational: blue-dominant rectangle-ish
    if feat["rect_fill"] >= RECT_FILL_MIN and blue_area > max(red_area, AREA_MIN_FRAC*roi_area):
        return "Informational", f"blue rectangle (fill={feat['rect_fill']:.2f})"

    # Fallbacks
    if feat["v"]==4 and blue_area > red_area:
        return "Informational", "v=4 (fallback) + blue-dominant"
    return "UNKNOWN", "no rule matched"

# ---------------- Full image pipeline ----------------
def detect_and_classify_all(img):
    H,W = img.shape[:2]
    red_mask, blue_mask = color_masks(img)
    min_area = 0.004 * H * W   # candidate min area (tune if needed)

    boxes = contours_from_mask(red_mask, min_area) + contours_from_mask(blue_mask, min_area)
    boxes = merge_boxes(boxes, iou_thr=0.2)
    rois = []
    for b in boxes:
        bx = pad_box(b, img.shape, pad=10)
        x,y,w,h = bx
        crop = img[y:y+h, x:x+w]
        label, rule = classify_roi(crop, (red_mask, blue_mask))
        rois.append({"box":bx, "crop":crop, "label":label, "rule":rule})
    return rois

def annotate_overview(img, rois):
    vis = img.copy()
    for r in rois:
        x,y,w,h = r["box"]
        label = r["label"]
        color = (0,255,0) if label in CATEGORIES else (0,0,255)
        cv.rectangle(vis,(x,y),(x+w,y+h),color,2)
        banner = f"{label} / {PERSIAN.get(label,label)}"
        cv.rectangle(vis,(x,max(0,y-26)),(x+340,y),(0,0,0),-1)
        cv.putText(vis,banner,(x+6,y-6),cv.FONT_HERSHEY_SIMPLEX,0.6,(255,255,255),2,cv.LINE_AA)
    return vis

# ---------------- Interactive upload (multi-sign image) ----------------
base = pathlib.Path("/content/outputs_multisign");
(base/"Warning").mkdir(parents=True, exist_ok=True)
(base/"Prohibitory").mkdir(parents=True, exist_ok=True)
(base/"Informational").mkdir(parents=True, exist_ok=True)
(base/"UNKNOWN").mkdir(parents=True, exist_ok=True)

print("Upload a multi-sign image (you can repeat). Crops will be saved per class in /content/outputs_multisign/*")

while True:
    up = files.upload()
    if not up:
        print("No file uploaded. Exiting."); break
    for name,data in up.items():
        p = f"/content/{name}"; open(p,"wb").write(data)
        img = cv.imread(p, cv.IMREAD_COLOR)
        rois = detect_and_classify_all(img)
        # save crops per class
        for i,r in enumerate(rois,1):
            out_dir = base / (r["label"] if r["label"] in CATEGORIES else "UNKNOWN")
            out_path = out_dir / f"{pathlib.Path(name).stem}_roi{i}.png"
            cv.imwrite(str(out_path), r["crop"])
        # overview image
        overview = annotate_overview(img, rois)
        out_over = base / f"{pathlib.Path(name).stem}_overview.png"
        cv.imwrite(str(out_over), overview)
        # show & report
        print(f"{name}: found {len(rois)} signs ->", {c:sum(r['label']==c for r in rois) for c in CATEGORIES})
        cv2_imshow(overview)
    if input("Again? (y/n): ").strip().lower() != "y":
        print("Done."); break
```

```python
# ==== GTSRB download to Colab/Drive with visible progress & auto-locate (dir or CSV) ====
# - Uses Kaggle CLI with progress bar if kaggle.json is present
# - Else uses kagglehub; shows tqdm bars for extraction/copy
# - Saves to Google Drive if desired
# - Locates either class-root (00000..00042) OR Train.csv fallback

!pip -q install kagglehub kaggle tqdm

import os, glob, zipfile, shutil, sys
from pathlib import Path
from tqdm.auto import tqdm

# ---------------- Settings ----------------
DATASET_ID = "meowmeowmeowmeowmeow/gtsrb-german-traffic-sign"

SAVE_TO_DRIVE = True   # <-       True 
DRIVE_DIR = "/content/drive/MyDrive/gtsrb_data"
LOCAL_DIR = "/content/gtsrb_data"

# -------------- Mount Drive (optional) --------------
if SAVE_TO_DRIVE:
    from google.colab import drive
    drive.mount("/content/drive", force_remount=True)
TARGET_DIR = DRIVE_DIR if SAVE_TO_DRIVE else LOCAL_DIR
os.makedirs(TARGET_DIR, exist_ok=True)

# -------------- Helpers --------------
def extract_zip_with_progress(zip_path: str, out_dir: str):
    with zipfile.ZipFile(zip_path, 'r') as zf:
        members = zf.infolist()
        pbar = tqdm(total=len(members), unit="file", desc=f"Extracting {Path(zip_path).name}")
        for m in members:
            try:
                zf.extract(m, out_dir)
            except Exception as e:
                print(f"[WARN] extract failed for {m.filename}: {e}")
            pbar.update(1)
        pbar.close()

def iter_files(root: str):
    for r, _, files in os.walk(root):
        for f in files:
            yield os.path.join(r, f)

def copytree_with_progress(src: str, dst: str, desc="Copying"):
    files = list(iter_files(src))
    if not files:
        print(f"[WARN] No files to copy in: {src}")
        return
    pbar = tqdm(total=len(files), unit="file", desc=f"{desc} -> {Path(dst).name}")
    for fp in files:
        rel = os.path.relpath(fp, src)
        outp = os.path.join(dst, rel)
        os.makedirs(os.path.dirname(outp), exist_ok=True)
        try:
            shutil.copy2(fp, outp)
        except Exception as e:
            print(f"[WARN] copy failed for {fp}: {e}")
        pbar.update(1)
    pbar.close()

def show_tree(path, levels=2):
    path = Path(path)
    print(f"\n[Tree @ {path}]")
    for root, dirs, files in os.walk(path):
        depth = Path(root).relative_to(path).parts
        if len(depth) > levels:
            continue
        indent = "  " * len(depth)
        print(f"{indent}{Path(root).name}/  (dirs: {len(dirs)}, files: {len(files)})")
        for d in sorted(dirs)[:10]:
            print(f"{indent}   {d}/")
        for f in sorted(files)[:10]:
            print(f"{indent}   {f}")

def auto_locate_class_root(base):
    """Find folder that directly contains many '00000'.. class dirs with images."""
    def looks_like_class_dir(p):
        b = os.path.basename(p)
        return b.isdigit() and len(b) == 5
    best = None
    for root, dirs, _ in os.walk(base):
        class_dirs = [d for d in dirs if looks_like_class_dir(os.path.join(root, d))]
        if len(class_dirs) >= 10:
            # ensure at least one class has images
            ok = False
            for d in class_dirs[:10]:
                for ext in ("*.ppm","*.png","*.jpg","*.jpeg","*.bmp"):
                    if glob.glob(os.path.join(root, d, ext)):
                        ok = True; break
                if ok: break
            if ok:
                best = root
                break
    # Common fallbacks
    if best is None:
        COMMON = [
            os.path.join(base, "GTSRB", "Final_Training", "Images"),
            os.path.join(base, "Final_Training", "Images"),
            os.path.join(base, "Training"),
            os.path.join(base, "Train"),
        ]
        for p in COMMON:
            if os.path.isdir(p):
                subs = [d for d in os.listdir(p) if os.path.isdir(os.path.join(p,d)) and d.isdigit() and len(d)==5]
                if len(subs) >= 10:
                    # optional: check images inside a few
                    best = p
                    break
    return best

def find_train_csv(base):
    """Search for a Train.csv / train.csv (fallback layout)."""
    candidates = []
    for root, _, files in os.walk(base):
        for f in files:
            if f.lower() in ("train.csv", "train_fixed.csv", "gt_train.csv"):
                candidates.append(os.path.join(root, f))
    return candidates[0] if candidates else None

# -------------- Download --------------
print("[INFO] Checking Kaggle credentials ...")
has_kaggle_token = os.path.exists("/root/.kaggle/kaggle.json")

if has_kaggle_token:
    print("[OK] Using Kaggle CLI (with progress bar).")
    try:
        os.chmod("/root/.kaggle/kaggle.json", 0o600)
    except Exception:
        pass
    from IPython import get_ipython
    # Download zip(s) into a temp dir inside TARGET_DIR
    ZIP_DIR = os.path.join(TARGET_DIR, "_zips")
    os.makedirs(ZIP_DIR, exist_ok=True)
    get_ipython().system(f'kaggle datasets download -d {DATASET_ID} -p "{ZIP_DIR}" -o')
    zips = glob.glob(os.path.join(ZIP_DIR, "*.zip"))
    if not zips:
        print("[WARN] No zip files found by Kaggle CLI. Showing ZIP_DIR tree:")
        show_tree(ZIP_DIR, levels=2)
    for zp in zips:
        extract_zip_with_progress(zp, TARGET_DIR)
else:
    print("[INFO] No kaggle.json found. Using kagglehub (download without bar); showing bars for extract/copy.")
    import kagglehub
    kh_path = kagglehub.dataset_download(DATASET_ID)
    print("KaggleHub cache path:", kh_path)
    # Extract any zips; if none, copy files with progress
    zip_files = glob.glob(os.path.join(kh_path, "**", "*.zip"), recursive=True)
    if zip_files:
        for zp in zip_files:
            extract_zip_with_progress(zp, TARGET_DIR)
    else:
        copytree_with_progress(kh_path, TARGET_DIR, desc="Copying from kagglehub cache")

# -------------- Locate class-root or CSV --------------
DATA_ROOT = auto_locate_class_root(TARGET_DIR)
TRAIN_CSV = find_train_csv(TARGET_DIR)

print("\n========== RESULT ==========")
print("TARGET_DIR:", TARGET_DIR)
print("DATA_ROOT :", DATA_ROOT)
print("TRAIN_CSV :", TRAIN_CSV)
if not DATA_ROOT and not TRAIN_CSV:
    print(" Could not locate class folders or CSV. Printing a short tree:")
    show_tree(TARGET_DIR, levels=3)
else:
    if DATA_ROOT:
        cls = sorted([d for d in os.listdir(DATA_ROOT) if d.isdigit() and len(d)==5])
        print("Found", len(cls), "class folders. Examples:", cls[:10])
    if TRAIN_CSV:
        print("Found training CSV:", TRAIN_CSV)
```

```python
# ===================== HOG-Only on EXACT SAME DATA SELECTION LOGIC =====================
# - No new folders are created. Uses your existing DATA_DIR as-is.
# - Class selection = EXACTLY like your ResNet cell:
#     * scan DATA_DIR for class folders
#     * sort with key=lambda x: int(x) if x.isdigit() else x
#     * pick FIRST 10 class names
# - Per-class image cap (MAX_PER_CLASS) and stratified split = same logic.
# - Classifier: HOG features + class prototypes (mean HOG) + cosine similarity (NO ML).
# - Outputs: per-split accuracy (train/val/test), plus test classification report & confusion matrix.

# If you already mounted Drive, you can comment out the next two lines.
from google.colab import drive
drive.mount('/content/drive', force_remount=True)

import os, re, math, numpy as np, cv2 as cv
from pathlib import Path
from tqdm.auto import tqdm
from collections import defaultdict
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

# -------------------- Config (MATCHES YOUR RESNET SELECTION BEHAVIOR) --------------------
DATA_DIR       = "/content/drive/MyDrive/gtsrb_data/TrainMain"  # <-- your dataset root (10 class folders)
MAX_PER_CLASS  = 200           # cap per class (set None to use ALL images)
SPLIT          = (0.70, 0.15, 0.15)  # train, val, test
SEED           = 42

# -------------------- Discover 10 classes (EXACT logic) --------------------
scan_dirs = [d for d in os.scandir(DATA_DIR) if d.is_dir()]
cls_all   = sorted([d.name for d in scan_dirs], key=lambda x: int(x) if x.isdigit() else x)
if len(cls_all) < 10:
    raise RuntimeError(f"Found only {len(cls_all)} class folders in {DATA_DIR}, need >= 10.")
cls_names = cls_all[:10]
print(f"[INFO] Selected classes ({len(cls_names)}): {cls_names}")

# Consistent class  index mapping (same order as cls_names)
class_to_idx = {c:i for i,c in enumerate(cls_names)}
n_classes = len(cls_names)

# -------------------- Gather file lists per class (shallow) --------------------
IMG_EXTS = (".ppm",".png",".jpg",".jpeg",".bmp",".tif",".tiff")
def list_images_shallow(folder: Path):
    return sorted([p for p in folder.iterdir() if p.is_file() and p.suffix.lower() in IMG_EXTS])

rng = np.random.default_rng(SEED)

per_class_files = {}
total_before_cap = 0
for cname in cls_names:
    cdir = Path(DATA_DIR) / cname
    if not cdir.is_dir():
        raise RuntimeError(f"Missing class folder: {cdir}")
    files = list_images_shallow(cdir)
    rng.shuffle(files)  # shuffle before capping/splitting (matches your intent)
    total_before_cap += len(files)
    if MAX_PER_CLASS is not None:
        files = files[:min(MAX_PER_CLASS, len(files))]
    per_class_files[cname] = files

print(f"Total images BEFORE cap: {total_before_cap}")
print(f"Total images AFTER  cap: {sum(len(v) for v in per_class_files.values())}")

# -------------------- Stratified split with per-class cap (EXACT style) --------------------
tr_r, va_r, te_r = SPLIT
assert abs((tr_r + va_r + te_r) - 1.0) < 1e-6

train_paths, val_paths, test_paths = [], [], []
train_labels, val_labels, test_labels = [], [], []

def _counts(labels, n_cls):
    arr = np.array(labels, dtype=int)
    return [int((arr == c).sum()) for c in range(n_cls)]

for cname, files in per_class_files.items():
    c = class_to_idx[cname]
    n = len(files)
    n_tr = int(round(tr_r * n))
    n_va = int(round(va_r * n))
    n_te = n - n_tr - n_va
    if n_te < 0:
        n_te = 0
        n_va = n - n_tr

    train_paths += files[:n_tr]
    val_paths   += files[n_tr:n_tr+n_va]
    test_paths  += files[n_tr+n_va:]

    train_labels += [c] * n_tr
    val_labels   += [c] * n_va
    test_labels  += [c] * n_te

print("Per-class counts AFTER CAP & SPLIT:")
print("  train:", _counts(train_labels, n_classes))
print("  val  :", _counts(val_labels,   n_classes))
print("  test :", _counts(test_labels,  n_classes))
print(f"Totals -> train={len(train_paths)}, val={len(val_paths)}, test={len(test_paths)}")

# -------------------- HOG descriptor (ONLY feature) --------------------
HOG_WIN = (64, 64)
_hog = cv.HOGDescriptor(
    _winSize=HOG_WIN,
    _blockSize=(16,16),
    _blockStride=(8,8),
    _cellSize=(8,8),
    _nbins=9
)

def read_img(path: Path):
    img = cv.imread(str(path), cv.IMREAD_COLOR)
    if img is None:
        raise RuntimeError(f"Cannot read: {path}")
    return img

def hog_features(img, size=HOG_WIN):
    img_rs = cv.resize(img, size, interpolation=cv.INTER_AREA)
    gray = cv.cvtColor(img_rs, cv.COLOR_BGR2GRAY)
    h = _hog.compute(gray).reshape(-1).astype(np.float32)
    n = np.linalg.norm(h) + 1e-12
    return h / n  # L2 normalization

def build_features(paths):
    X = []
    for p in tqdm(paths, desc="HOG", unit="img", leave=False):
        im = read_img(Path(p))
        X.append(hog_features(im))
    return np.vstack(X).astype(np.float32) if len(X) else np.zeros((0, _hog.compute(np.zeros(HOG_WIN, np.uint8)).size), np.float32)

# -------------------- Extract HOG for each split --------------------
print("\n[INFO] Extracting HOG features ...")
X_train = build_features(train_paths)
X_val   = build_features(val_paths)
X_test  = build_features(test_paths)

y_train = np.array(train_labels, dtype=np.int32)
y_val   = np.array(val_labels,   dtype=np.int32)
y_test  = np.array(test_labels,  dtype=np.int32)

print("Shapes ->",
      f"X_train: {X_train.shape}, y_train: {y_train.shape} |",
      f"X_val: {X_val.shape}, y_val: {y_val.shape} |",
      f"X_test: {X_test.shape}, y_test: {y_test.shape}")

# -------------------- Build class prototypes on TRAIN ONLY --------------------
def compute_prototypes(X, y, n_cls):
    protos = np.zeros((n_cls, X.shape[1]), dtype=np.float32)
    has = np.zeros(n_cls, dtype=bool)
    for c in range(n_cls):
        idx = (y == c)
        if idx.any():
            v = X[idx].mean(axis=0)
            v /= (np.linalg.norm(v) + 1e-12)
            protos[c] = v
            has[c] = True
    # For classes with zero train samples (edge case), leave prototype as zeros.
    # They simply wont be predicted unless similarity defaults.
    return protos, has

protos, proto_mask = compute_prototypes(X_train, y_train, n_classes)

# -------------------- Cosine-similarity classifier (NO ML) --------------------
def predict_batch(X, protos):
    if X.size == 0:
        return np.array([], dtype=np.int32)
    Xn = X / (np.linalg.norm(X, axis=1, keepdims=True) + 1e-12)
    sims = Xn @ protos.T                       # (N x D) @ (D x C) -> (N x C)
    idx = np.argmax(sims, axis=1)
    return idx.astype(np.int32)

# -------------------- Evaluate --------------------
def evaluate_split(name, X, y_true):
    if len(y_true) == 0:
        print(f"[{name}] No samples.")
        return None, None
    y_pred = predict_batch(X, protos)
    acc = accuracy_score(y_true, y_pred)
    print(f"[{name}] Accuracy: {acc*100:.2f}% (n={len(y_true)})")
    return y_pred, acc

print("\n[INFO] Evaluating ...")
_yp_tr, acc_tr = evaluate_split("TRAIN", X_train, y_train)
_yp_va, acc_va = evaluate_split("VAL",   X_val,   y_val)
y_pred_test, acc_te = evaluate_split("TEST",  X_test,  y_test)

# Detailed report for TEST only (if multiple classes present)
if y_pred_test is not None and len(np.unique(y_test)) > 1:
    # Use numeric labels as class names to avoid mismatch if some classes absent in test
    print("\n=== TEST Classification Report ===")
    print(classification_report(y_test, y_pred_test, target_names=[str(i) for i in range(n_classes)]))
    print("=== TEST Confusion Matrix (rows=true, cols=pred) ===")
    print(confusion_matrix(y_test, y_pred_test, labels=list(range(n_classes))))

# -------------------- Quick single-image predictor --------------------
def predict_image(img_path: str):
    """
    Predict one image path using TRAIN prototypes.
    Returns: (pred_idx, class_name)
    """
    img = read_img(Path(img_path))
    feat = hog_features(img).reshape(1, -1)
    pred = int(predict_batch(feat, protos)[0])
    cname = cls_names[pred] if 0 <= pred < len(cls_names) else str(pred)
    print(f"Predicted: {pred} ({cname})")
    return pred, cname

# Example (uncomment to test):
# _ = predict_image(str(test_paths[0]))
```

```python
# ===================== One Colab Cell: Split + Train + Log + Plot + Test =====================
# Works on Colab (A100 High-RAM). Shows per-epoch metrics and saves them to CSV.

# 0) Setup & Mount Drive
from google.colab import drive
drive.mount('/content/drive', force_remount=True)

import os, random, time, json
import numpy as np
import pandas as pd
import torch
from torch import nn
from torch.utils.data import DataLoader, Subset
from torchvision import datasets, transforms
from torchvision.models import resnet18, ResNet18_Weights
import matplotlib.pyplot as plt
from tqdm.auto import tqdm

# -------------------- Config --------------------
DATA_DIR   = "/content/drive/MyDrive/gtsrb_data/TrainMain"  # <-- your dataset root (10 class folders)
IMG_SIZE   = 224
BATCH_SIZE = 64
EPOCHS     = 15
LR         = 1e-3
WEIGHT_DECAY = 1e-4
SPLIT = (0.70, 0.15, 0.15)      # train, val, test
MAX_PER_CLASS = 200            # cap per class (set None to use all images)
SEED = 42
NUM_WORKERS = 0                 # if Drive I/O is slow, set to 0

SAVE_DIR = "/content/drive/MyDrive/gtsrb_data"
os.makedirs(SAVE_DIR, exist_ok=True)
SAVE_PATH = os.path.join(SAVE_DIR, "resnet18_best.pt")
METRICS_CSV = os.path.join(SAVE_DIR, "metrics.csv")
CLASSMAP_JSON = os.path.join(SAVE_DIR, "classmap.json")

# -------------------- Reproducibility --------------------
random.seed(SEED)
np.random.seed(SEED)
torch.manual_seed(SEED)
torch.cuda.manual_seed_all(SEED)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Device:", device)
if device.type == "cuda":
    try:
        torch.backends.cudnn.benchmark = True
        torch.set_float32_matmul_precision('high')
    except Exception as e:
        print("Note:", e)

# -------------------- Transforms --------------------
mean_imagenet = [0.485, 0.456, 0.406]
std_imagenet  = [0.229, 0.224, 0.225]

train_tfms = transforms.Compose([
    transforms.Resize((IMG_SIZE, IMG_SIZE)),
    transforms.RandomHorizontalFlip(0.5),
    transforms.RandomRotation(10),
    transforms.ColorJitter(brightness=0.2, contrast=0.2),
    transforms.ToTensor(),
    transforms.Normalize(mean_imagenet, std_imagenet),
])

eval_tfms = transforms.Compose([
    transforms.Resize((IMG_SIZE, IMG_SIZE)),
    transforms.ToTensor(),
    transforms.Normalize(mean_imagenet, std_imagenet),
])

# -------------------- Filtered ImageFolder --------------------
class FilteredImageFolder(datasets.ImageFolder):
    def __init__(self, root, selected_classnames, transform=None):
        super().__init__(root=root, transform=transform)
        self.class_to_idx = {c:i for i,c in enumerate(selected_classnames)}
        new_samples = []
        for p, _ in self.samples:
            clsname = os.path.basename(os.path.dirname(p))
            if clsname in self.class_to_idx:
                new_samples.append((p, self.class_to_idx[clsname]))
        self.samples = new_samples
        self.imgs = self.samples
        self.targets = [y for _, y in self.samples]
        self.classes = selected_classnames

# -------------------- Discover 10 classes --------------------
cls_all = sorted([d.name for d in os.scandir(DATA_DIR) if d.is_dir()],
                 key=lambda x: int(x) if x.isdigit() else x)
if len(cls_all) < 10:
    raise RuntimeError(f"Found only {len(cls_all)} class folders, need 10.")
cls_names = cls_all[:10]
print(f"Using classes ({len(cls_names)}): {cls_names}")

# Build datasets
all_for_split = FilteredImageFolder(DATA_DIR, cls_names, transform=None)
train_base = FilteredImageFolder(DATA_DIR, cls_names, transform=train_tfms)
val_base   = FilteredImageFolder(DATA_DIR, cls_names, transform=eval_tfms)
test_base  = FilteredImageFolder(DATA_DIR, cls_names, transform=eval_tfms)

labels = np.array(all_for_split.targets)
n_classes = len(cls_names)
print("Total images (before cap):", len(labels))

# -------------------- Stratified split with per-class cap --------------------
tr_r, va_r, te_r = SPLIT
assert abs((tr_r+va_r+te_r)-1.0) < 1e-6

train_idx, val_idx, test_idx = [], [], []
rng = np.random.default_rng(SEED)

for c in range(n_classes):
    idx_all = np.where(labels == c)[0]
    rng.shuffle(idx_all)
    if MAX_PER_CLASS is not None:
        idx = idx_all[:min(MAX_PER_CLASS, len(idx_all))]
    else:
        idx = idx_all

    n = len(idx)
    n_tr = int(round(tr_r*n))
    n_va = int(round(va_r*n))
    n_te = n - n_tr - n_va
    if n_te < 0:
        n_te = 0
        n_va = n - n_tr

    train_idx += idx[:n_tr].tolist()
    val_idx   += idx[n_tr:n_tr+n_va].tolist()
    test_idx  += idx[n_tr+n_va:].tolist()

def counts(ix):
    return [int((labels[ix]==c).sum()) for c in range(n_classes)]

print("Per-class counts AFTER CAP:")
print("  train:", counts(np.array(train_idx)))
print("  val  :", counts(np.array(val_idx)))
print("  test :", counts(np.array(test_idx)))
print(f"Totals -> train={len(train_idx)}, val={len(val_idx)}, test={len(test_idx)}")

train_ds = Subset(train_base, train_idx)
val_ds   = Subset(val_base,   val_idx)
test_ds  = Subset(test_base,  test_idx)

train_loader = DataLoader(train_ds, batch_size=BATCH_SIZE, shuffle=True,
                          num_workers=NUM_WORKERS, pin_memory=(device.type=="cuda"))
val_loader   = DataLoader(val_ds,   batch_size=BATCH_SIZE, shuffle=False,
                          num_workers=NUM_WORKERS, pin_memory=(device.type=="cuda"))
test_loader  = DataLoader(test_ds,  batch_size=BATCH_SIZE, shuffle=False,
                          num_workers=NUM_WORKERS, pin_memory=(device.type=="cuda"))

# -------------------- Model --------------------
weights = ResNet18_Weights.DEFAULT
model = resnet18(weights=weights)
model.fc = nn.Linear(model.fc.in_features, n_classes)
model = model.to(device)

criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.AdamW(model.parameters(), lr=LR, weight_decay=WEIGHT_DECAY)
scheduler = torch.optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=EPOCHS)
scaler = torch.cuda.amp.GradScaler(enabled=(device.type=="cuda"))

# -------------------- Train / Eval helpers (with tqdm) --------------------
def run_epoch(loader, train_mode=True, desc="Train"):
    if train_mode:
        model.train()
    else:
        model.eval()
    total_loss, total_correct, total_n = 0.0, 0, 0
    iterator = tqdm(loader, desc=desc, leave=False, dynamic_ncols=True)

    if train_mode:
        for x, y in iterator:
            x, y = x.to(device, non_blocking=True), y.to(device, non_blocking=True)
            optimizer.zero_grad(set_to_none=True)
            with torch.cuda.amp.autocast(enabled=(device.type=="cuda")):
                logits = model(x)
                loss = criterion(logits, y)
            scaler.scale(loss).backward()
            scaler.step(optimizer)
            scaler.update()

            total_loss   += loss.item() * x.size(0)
            total_correct += (logits.argmax(1) == y).sum().item()
            total_n      += x.size(0)
            iterator.set_postfix(loss=f"{loss.item():.4f}")
    else:
        with torch.no_grad():
            for x, y in iterator:
                x, y = x.to(device, non_blocking=True), y.to(device, non_blocking=True)
                logits = model(x)
                loss = criterion(logits, y)
                total_loss   += loss.item() * x.size(0)
                total_correct += (logits.argmax(1) == y).sum().item()
                total_n      += x.size(0)
                iterator.set_postfix(loss=f"{loss.item():.4f}")

    return total_loss / total_n, total_correct / total_n

# -------------------- Training loop (prints per-epoch metrics & saves CSV) --------------------
best_val_acc, best_state = 0.0, None
train_losses, val_losses, train_accs, val_accs = [], [], [], []
metrics_log = []  # will save each epochs metrics

print("\nStarting training ...\n")
for epoch in tqdm(range(1, EPOCHS + 1), desc="Epochs", dynamic_ncols=True):
    tr_loss, tr_acc = run_epoch(train_loader, train_mode=True,  desc=f"Train {epoch}/{EPOCHS}")
    val_loss, val_acc = run_epoch(val_loader,   train_mode=False, desc=f"Val   {epoch}/{EPOCHS}")
    scheduler.step()

    train_losses.append(tr_loss); train_accs.append(tr_acc)
    val_losses.append(val_loss);   val_accs.append(val_acc)

    # ---- Print per-epoch summary (Accuracy & Loss) ----
    tqdm.write(f"Epoch {epoch:02d} | "
               f"train_loss={tr_loss:.4f}  train_acc={tr_acc*100:.2f}% | "
               f"val_loss={val_loss:.4f}    val_acc={val_acc*100:.2f}%")

    # ---- Record metrics for CSV ----
    current_lr = optimizer.param_groups[0]["lr"]
    metrics_log.append({
        "epoch": epoch,
        "train_loss": float(tr_loss),
        "train_acc": float(tr_acc),
        "val_loss": float(val_loss),
        "val_acc": float(val_acc),
        "lr": float(current_lr),
    })

    # Track best model by val_acc
    if val_acc > best_val_acc:
        best_val_acc = val_acc
        best_state = {k: v.cpu() for k, v in model.state_dict().items()}

# Save best checkpoint
if best_state is not None:
    model.load_state_dict(best_state, strict=True)
    torch.save(model.state_dict(), SAVE_PATH)
    print(f"\nBest model (val_acc={best_val_acc*100:.2f}%) saved to: {SAVE_PATH}")

# Save metrics to CSV
df = pd.DataFrame(metrics_log, columns=["epoch","train_loss","train_acc","val_loss","val_acc","lr"])
df.to_csv(METRICS_CSV, index=False)
print(f"\nEpoch metrics saved to: {METRICS_CSV}")
print(df.tail(5))  # preview last 5 rows

# Also save class  index mapping
classmap = {i: c for i, c in enumerate(cls_names)}
with open(CLASSMAP_JSON, "w") as f:
    json.dump(classmap, f, indent=2)
print(f"Class map saved to: {CLASSMAP_JSON}")

# -------------------- Test set evaluation --------------------
model.eval()
test_correct, test_n = 0, 0
with torch.no_grad():
    for x, y in test_loader:
        x, y = x.to(device, non_blocking=True), y.to(device, non_blocking=True)
        logits = model(x)
        test_correct += (logits.argmax(1) == y).sum().item()
        test_n += x.size(0)
test_acc = test_correct / max(1, test_n)
print(f"\nTEST ACCURACY: {test_acc*100:.2f}% on {test_n} images")

# -------------------- Plots --------------------
epochs = range(1, EPOCHS+1)
plt.figure(figsize=(6,4))
plt.plot(epochs, [a*100 for a in train_accs], label="Train Acc (%)")
plt.plot(epochs, [a*100 for a in val_accs],   label="Val Acc (%)")
plt.xlabel("Epoch"); plt.ylabel("Accuracy (%)"); plt.title("Accuracy vs Epoch"); plt.legend(); plt.grid(True)
plt.show()

plt.figure(figsize=(6,4))
plt.plot(epochs, train_losses, label="Train Loss")
plt.plot(epochs, val_losses,   label="Val Loss")
plt.xlabel("Epoch"); plt.ylabel("Loss"); plt.title("Loss vs Epoch"); plt.legend(); plt.grid(True)
plt.show()

print("\nClass-to-index mapping:")
print(classmap)
# ===================== End Cell =====================
```

```python
# =========================
# Colab + Google Drive setup
# =========================
from google.colab import drive
drive.mount('/content/drive')

import os, cv2, math, glob
import numpy as np
import matplotlib.pyplot as plt

#     
ROOT = "/content/drive/MyDrive/Q2"
SCENE_PATH = os.path.join(ROOT, "Multimedia-HW3-Q2-pic-1.jpg")

#  (   )
OUT_DIR   = os.path.join(ROOT, "outputs")
MASKS_DIR = os.path.join(OUT_DIR, "masks")
CROPS_DIR = os.path.join(OUT_DIR, "objects_cropped")
CAT_DIR   = os.path.join(OUT_DIR, "category_outputs")
for d in [OUT_DIR, MASKS_DIR, CROPS_DIR, CAT_DIR]:
    os.makedirs(d, exist_ok=True)

# ---------- 33 grid helpers (  ) ----------
def fit_into_square(img_bgr, size):
    h, w = img_bgr.shape[:2]
    if h == 0 or w == 0:
        return 255*np.ones((size, size, 3), np.uint8)
    s = min(size/max(1,w), size/max(1,h))
    nw, nh = max(1,int(round(w*s))), max(1,int(round(h*s)))
    r = cv2.resize(img_bgr, (nw, nh), interpolation=cv2.INTER_AREA)
    canvas = 255*np.ones((size, size, 3), np.uint8)
    y0, x0 = (size-nh)//2, (size-nw)//2
    canvas[y0:y0+nh, x0:x0+nw] = r
    return canvas

def make_3x3_class_grid(class_to_crops, class_order, save_path):
    GRID_ROWS, GRID_COLS = 3, 3
    CELL_W, CELL_H = 600, 420
    BORDER, TITLE_H, PAD = 10, 48, 12

    W = GRID_COLS*CELL_W + (GRID_COLS+1)*BORDER
    H = GRID_ROWS*CELL_H + (GRID_ROWS+1)*BORDER
    canvas = 255*np.ones((H, W, 3), np.uint8)

    classes = list(class_order)[:GRID_ROWS*GRID_COLS]
    while len(classes) < GRID_ROWS*GRID_COLS:
        classes.append("")

    for idx, cat in enumerate(classes):
        r, c = divmod(idx, GRID_COLS)
        x0 = BORDER + c*(CELL_W + BORDER)
        y0 = BORDER + r*(CELL_H + BORDER)
        x1, y1 = x0 + CELL_W - 1, y0 + CELL_H - 1

        #  
        cv2.rectangle(canvas, (x0, y0), (x1, y1), (0,0,0), 3)

        #  
        title = cat if cat else ""
        (tw, th), _ = cv2.getTextSize(title, cv2.FONT_HERSHEY_SIMPLEX, 1.0, 2)
        tx, ty = x0 + (CELL_W - tw)//2, y0 + TITLE_H//2 + th//2 - 4
        cv2.putText(canvas, title, (tx, ty), cv2.FONT_HERSHEY_SIMPLEX, 1.0, (0,0,0), 2, cv2.LINE_AA)

        #  
        cx0, cy0 = x0 + PAD, y0 + TITLE_H + PAD
        cw, ch   = CELL_W - 2*PAD, CELL_H - TITLE_H - 2*PAD
        cv2.rectangle(canvas, (cx0, cy0), (cx0+cw-1, cy0+ch-1), (230,230,230), 1)

        #  
        thumbs = class_to_crops.get(cat, [])
        if not thumbs:
            continue
        GAP, THUMB = 8, 110
        cols_fit = max(1, (cw + GAP) // (THUMB + GAP))
        rows_fit = max(1, (ch + GAP) // (THUMB + GAP))
        slots = int(cols_fit*rows_fit)

        for i, img in enumerate(thumbs[:slots]):
            rr, cc = i // cols_fit, i % cols_fit
            px = cx0 + GAP + cc*(THUMB + GAP)
            py = cy0 + GAP + rr*(THUMB + GAP)
            t = fit_into_square(img, THUMB)
            h, w = t.shape[:2]
            if py+h > cy0+ch: break
            if px+w > cx0+cw: continue
            canvas[py:py+h, px:px+w] = t

        extra = len(thumbs) - slots
        if extra > 0:
            msg = f"+{extra} more"
            (mw, mh), _ = cv2.getTextSize(msg, cv2.FONT_HERSHEY_SIMPLEX, 0.7, 2)
            mx, my = cx0 + cw - mw - 6, cy0 + ch - 6
            cv2.putText(canvas, msg, (mx, my), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (80,80,80), 2, cv2.LINE_AA)

    cv2.imwrite(save_path, canvas)
    plt.figure(figsize=(16,10))
    plt.imshow(cv2.cvtColor(canvas, cv2.COLOR_BGR2RGB))
    plt.axis('off'); plt.title("33 Class Grid"); plt.show()

# =========================
# 1) Load scene (  )
# =========================
img = cv2.imread(SCENE_PATH)
if img is None:
    raise FileNotFoundError(f"Could not load image: {SCENE_PATH}")
orig = img.copy()

# =========================
# 2) Binary mask (   )
# =========================
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
_, thresh = cv2.threshold(gray, 240, 255, cv2.THRESH_BINARY_INV)
kernel = np.ones((15, 15), np.uint8)
thresh = cv2.dilate(thresh, kernel, iterations=1)
thresh = cv2.morphologyEx(thresh, cv2.MORPH_OPEN, np.ones((5, 5), np.uint8))

cv2.imwrite(os.path.join(MASKS_DIR, "full_binary_mask.png"), thresh)

# =========================
# 3) Find contours + debug view ()
# =========================
contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
print(f"Total contours found: {len(contours)}")

debug_img = orig.copy()
cv2.drawContours(debug_img, contours, -1, (0, 0, 255), 3)
plt.figure(figsize=(14,6))
plt.subplot(1,2,1); plt.imshow(cv2.cvtColor(debug_img, cv2.COLOR_BGR2RGB)); plt.title("Detected Contours"); plt.axis("off")
plt.subplot(1,2,2); plt.imshow(thresh, cmap="gray"); plt.title("Binary Mask"); plt.axis("off")
plt.tight_layout(); plt.show()

# =========================
# 4) Crop & clean ()
# =========================
object_images, object_contours, object_bboxes = [], [], []
for idx, cnt in enumerate(contours):
    x, y, w, h = cv2.boundingRect(cnt); object_bboxes.append((x,y,w,h))
    cnt_shifted = cnt - [x, y]
    mask = np.zeros((h, w), dtype=np.uint8)
    cv2.drawContours(mask, [cnt_shifted], -1, 255, -1)
    cropped = orig[y:y+h, x:x+w]

    gray_crop = cv2.cvtColor(cropped, cv2.COLOR_BGR2GRAY)
    _, bin_mask = cv2.threshold(gray_crop, 240, 255, cv2.THRESH_BINARY_INV)
    k3 = np.ones((3,3), np.uint8)
    bin_mask = cv2.morphologyEx(bin_mask, cv2.MORPH_OPEN, k3, iterations=1)
    bin_mask = cv2.erode(bin_mask, k3, iterations=1)

    clean = cv2.bitwise_and(cropped, cropped, mask=bin_mask)
    object_images.append(clean); object_contours.append(cnt)
    cv2.imwrite(os.path.join(CROPS_DIR, f"obj_{idx+1}.png"), clean)

print(f"Saved {len(object_images)} crops in '{CROPS_DIR}'.")

# =========================
# 5) Classification   :     + min-3
# =========================
# :        .  1-Base  (  ).
category_reps_1based = {
    "Number_7": 25,
    "Number_8": 12,
    "Number_9": 20,
    "Apple":    16,
    "Cucumber": 19,
    "Banana":   15,
    "Deer":     18,
    "Lion":     10,
    "Elephant": 7
}
#  0-Base  
category_reps = {k: v-1 for k, v in category_reps_1based.items()}

categories = {cat: [] for cat in category_reps}  # cat -> list[(obj_idx, score)]
assigned = {}

#  :      
for idx, cnt in enumerate(object_contours):
    best_cat, best_score = None, float("inf")
    for cat, ref_idx in category_reps.items():
        ref_cnt = object_contours[ref_idx]
        s = cv2.matchShapes(cnt, ref_cnt, cv2.CONTOURS_MATCH_I1, 0.0)
        if s < best_score:
            best_score, best_cat = s, cat
    categories[best_cat].append((idx, best_score))
    assigned[idx] = best_cat

#         (  )
for cat, items in categories.items():
    while len(items) < 3:
        best_candidate, best_score = None, float("inf")
        for other_cat, other_items in categories.items():
            if other_cat == cat or len(other_items) <= 3:
                continue
            for obj_idx, _ in other_items:
                s = cv2.matchShapes(object_contours[obj_idx],
                                    object_contours[category_reps[cat]],
                                    cv2.CONTOURS_MATCH_I1, 0.0)
                if s < best_score:
                    best_score = s
                    best_candidate = (obj_idx, best_score, other_cat)
        if best_candidate is None:
            break
        obj_idx, s, from_cat = best_candidate
        categories[from_cat] = [(i, sc) for (i, sc) in categories[from_cat] if i != obj_idx]
        categories[cat].append((obj_idx, s))
        assigned[obj_idx] = cat

#  
for cat, items in categories.items():
    print(f"\nCategory: {cat} (total: {len(items)})")
    for idx, sc in items:
        print(f" - obj_{idx+1}.png | score={sc:.4f}")

# =========================
# 6)      + 
# =========================
H, W = orig.shape[:2]
per_cat_canvas = {cat: np.full((H, W, 3), 255, np.uint8) for cat in categories.keys()}
overlay = orig.copy()

for cat, items in categories.items():
    for idx, _ in items:
        cnt = object_contours[idx]
        mask_full = np.zeros((H, W), dtype=np.uint8)
        cv2.drawContours(mask_full, [cnt], -1, 255, cv2.FILLED)
        per_cat_canvas[cat][mask_full==255] = orig[mask_full==255]

for cat, canvas in per_cat_canvas.items():
    cv2.imwrite(os.path.join(CAT_DIR, f"{os.path.splitext(os.path.basename(SCENE_PATH))[0]}_{cat}.png"), canvas)

#    
for idx, cnt in enumerate(object_contours):
    x,y,w,h = cv2.boundingRect(cnt)
    cv2.rectangle(overlay, (x,y), (x+w,y+h), (0,0,255), 2)
    cv2.putText(overlay, assigned[idx], (x, y-5), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0,0,255), 1, cv2.LINE_AA)
cv2.imwrite(os.path.join(OUT_DIR, f"{os.path.splitext(os.path.basename(SCENE_PATH))[0]}_overlay.png"), overlay)

# =========================
# 7)   33 (   )
# =========================
class_to_crops = {cat: [] for cat in categories.keys()}
for cat, items in categories.items():
    for idx, _ in items:
        crop_path = os.path.join(CROPS_DIR, f"obj_{idx+1}.png")
        img = cv2.imread(crop_path)
        if img is not None:
            class_to_crops[cat].append(img)

class_order = ["Apple","Banana","Cucumber","Deer","Elephant","Lion","Number_7","Number_8","Number_9"]
grid_path = os.path.join(OUT_DIR, f"{os.path.splitext(os.path.basename(SCENE_PATH))[0]}_3x3_grid.png")
make_3x3_class_grid(class_to_crops, class_order, grid_path)

#  
plt.figure(figsize=(12,5))
plt.subplot(1,2,1); plt.imshow(cv2.cvtColor(orig, cv2.COLOR_BGR2RGB)); plt.axis('off'); plt.title("Original")
plt.subplot(1,2,2); plt.imshow(cv2.cvtColor(overlay, cv2.COLOR_BGR2RGB)); plt.axis('off'); plt.title("Detections + labels")
plt.tight_layout(); plt.show()

print("\nDone.")
print("Overlay ->", os.path.join(OUT_DIR, f"{os.path.splitext(os.path.basename(SCENE_PATH))[0]}_overlay.png"))
print("3x3 grid ->", grid_path)
```
